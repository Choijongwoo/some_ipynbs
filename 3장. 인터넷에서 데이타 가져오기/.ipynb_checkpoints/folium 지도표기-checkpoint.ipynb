```python
import folium
import numpy as np
import pandas as pd
import googlemaps
from tqdm import tqdm_notebook
import re
```

```python
# df.to_csv('../data/03.crawl_sandwich_restaurants.csv', sep=',', encoding='utf-8')
df = pd.read_csv('../data/03.crawl_sandwich_restaurants.csv', index_col=0)
df.head()
```

```python
gmap_key = open('../data/googlemaps_api_key.txt').read()
```

```python
gmaps = googlemaps.Client(key=gmap_key)
```

```python
Addresses = []
for i in df.Addresses:
    if i != 'Multiple':
        Addresses.append(i + ' ' + 'Chicago')
    else:
        Addresses.append(i)
df.Addresses = Addresses
```

```python
lat = []
lng = []

for i in tqdm_notebook(df.index):
    if df.Addresses[i] != 'Multiple':
        lat.append(gmaps.geocode(df.Addresses[i])[0]['geometry']['location']['lat'])
        lng.append(gmaps.geocode(df.Addresses[i])[0]['geometry']['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
df['lat'] = lat
df['lng'] = lng
```

```python
df
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)
for i in df.index:
    if df.Addresses[i] != 'Multiple':
        folium.Marker([df['lat'][i], df['lng'][i]], popup=df['Cafe'][i]).add_to(mapping)
mapping
```
