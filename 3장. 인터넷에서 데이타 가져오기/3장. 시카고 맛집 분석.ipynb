```python
from bs4 import BeautifulSoup


import os
import re
```

```python
import requests
```

```python
url_base = 'http://www.chicagomag.com'
url_sub = '/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'
# http://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/
response = requests.get(url_base+url_sub)
html = response.text
soup = BeautifulSoup(html, 'html.parser')

tmp_sammy = soup.find_all('div',"sammy")
```

```python
tmp_sammylisting = soup.find_all('div',"sammyListing")
```

```python
sammyRank = []
url = []
menu = []
name = []
for i in tmp_sammy:
    sammyRank.append(i.find('div', 'sammyRank').get_text())
    url.append(i.find('a')['href'])
    A = i.find('div', 'sammyListing').get_text()
    menu.append(re.split("\n|\r\n", A)[0])
    name.append(re.split("\n|\r\n", A)[1])
```

```python
import pandas as pd
```

```python
data = {"sammyRank" : sammyRank, "name" : name, "menu" : menu, 'url' : url}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("../data/03.내가 얻어낸 맛집 데이타.csv", sep=',')
```

```python
from tqdm import tqdm_notebook
```

```python
texts = []
for i in tqdm_notebook(df.index):
    response = requests.get(df.URL[i])
    html = response.text
    soup = BeautifulSoup(html,'html.parser')
    text = soup.find('p', 'addy').find('em').get_text()
    texts.append(text)
```

```python
df['texts'] = texts
```

```python
df.to_csv('../data/03.내가 얻어낸 맛집 데이타.csv', sep=',')
```

```python
c = re.compile('\d+')
re.search(c,'$10.')[0]
```

```python
Prices = []
Addresses = []
for text in df.texts:
    i = text.split()
    price = i[0]
    Prices.append(re.search(c,price)[0])
    Addresses.append(' '.join(i[1:-2]))
```

```python
df['Prices'] = Prices
df['Addresses'] = Addresses
```
