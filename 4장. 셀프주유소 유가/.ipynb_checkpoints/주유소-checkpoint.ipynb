```python
gasoline_prices_by_gu = glob("../data/gasoline/gasoline*")
```

```python
len(gasoline_prices_by_gu)
```

```python
gasoline_prices_by_gu
```

```python
all_excels_LIST = []
for i in gasoline_prices_by_gu:

    all_excels_LIST.append(pd.read_excel(i, header = 2))
gasoline_price_all_in_one = pd.concat(all_excels_LIST)
```

```python
gasoline_price_all_in_one.reset_index(inplace=True)
```

```python
gasoline_price_all_in_one.to_excel("../data/gasoline_price_all_in_one.xls", encoding="utf-8")
```

```python
gasoline_price_all_in_one
```

```python
gasoline_prices = pd.DataFrame({'주유소':gasoline_price_all_in_one['상호'],
                                '위치':gasoline_price_all_in_one['주소'],
                                '셀프':gasoline_price_all_in_one['셀프여부'],
                                '가격':gasoline_price_all_in_one['휘발유'],
                                '상표':gasoline_price_all_in_one['상표']
                               })
gasoline_prices
```

```python
gasoline_prices['구'] = [i.split()[1]   for i in gasoline_prices['위치']]
```

```python
gasoline_prices
```

```python
gasoline_prices['구'].unique()
```

```python
gasoline_prices = gasoline_prices[gasoline_prices['가격'] != '-']
```

```python
gasoline_prices['가격'] = [float(i) for i in gasoline_prices['가격']]
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x = '상표', y = '가격', hue='셀프', data=gasoline_prices, palette='Set3')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.boxplot(x = '상표',y = '가격', hue ='셀프', data = gasoline_prices, palette ='Set3')
sns.swarmplot(x = '상표',y = '가격', data = gasoline_prices, color='.6')
plt.show()
```

```python
# 02. skorea_municipalities_geo_simple.json


geo_path = '../data/02. skorea_municipalities_geo_simple.json'
with open(geo_path, encoding='utf-8') as geodata:
    geodata = json.load(geodata)
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5, tiles="Stamen Toner")
map
```

```python
map.choropleth(geo_data=geodata, 
               data=pivot_gasoline_prices,
               columns= [pivot_gasoline_prices.index, '가격'],
               fill_color = "PuRd",
               key_on = "feature.id"
              )
map
```
