```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,50)
```

```python
y1 = x**3
y2 = x**2
```

```python
plt.figure()
plt.plot(x,y1)
plt.show()

plt.figure()
plt.plot(x,y2)
plt.show()
```

```python
new_ticks = np.linspace(-3,3,5)
new_ticks
```

```python
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2, color = "purple", linewidth = 2 , linestyle = "--")
plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.xlabel('this is X ------------->')
plt.ylabel('this is Y ------------->')
plt.xticks(new_ticks)
plt.yticks([-3,-2.2,-1,0,1,3.5],["very good", "not very good" , "good" , "maybe good", "bad" , "very bad"])
plt.show()
```

```python
plt.figure()


plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.xlabel('this is X ------------->')
plt.ylabel('this is Y ------------->')
plt.xticks(new_ticks)
plt.yticks([-3,-2.2,-1,0,1,3.5],
           ["very good", "not very good" , "good" , "maybe good", "bad" , "very bad"])

ax = plt.gca()
ax.spines
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")

ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")

ax.spines["bottom"].set_position(("data", 0))
ax.spines["left"].set_position(("data", -5))


plt.plot(x,y1,label = "up")
plt.plot(x,y2, color = "purple", linewidth = 2 , linestyle = "--", label = "down")


plt.legend(loc = "best")


plt.show()
```
