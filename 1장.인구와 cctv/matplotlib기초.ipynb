```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
### plt.figure(figsize = (5,5))



# lines
x = np.linspace(0,10,20)
y = x**2
plt.plot(x,y,label = 'red line',color = 'purple', lw = 1, linestyle = '--')
plt.scatter(x,y,label = 'blue line',color = "blue", lw =1)




# x & y's limit
plt.xlim()
plt.ylim(0,15)

# 刻度ticks
plt.xticks(np.linspace(0,10,11))
# plt.yticks([0,5,10,15],['zero','bad','good','very good'])
plt.yticks(np.linspace(0,10,11))

# 변두리 조작
axes = plt.gca()
axes.spines['top'].set_color('none')
axes.spines['right'].set_color('none')

axes.spines['left'].set_position(("data",0))
axes.spines['bottom'].set_position(("data",0))



# 레젼드 설정
plt.legend(labels = ['red line','blue line','hello' ],loc = 'best')



# 注释 annotation 

x0 = 3
y0 = x0**2


plt.scatter(x0,y0, color = "r", lw = 20)
plt.plot([x0,x0], [0,y0],'k--',lw = 2)



plt.scatter(x0,x0, color = "yellow", lw = 10)

plt.scatter(y0,0, color = "k", lw = 10)

plt.annotate("hello world, this is a annotation made for tell you this is a annotation", 
             xy = (3,5), xycoords = "data" , xytext = (30,-30), textcoords = "offset points" , 
             arrowprops = dict(arrowstyle = '->' , connectionstyle = 'arc3, rad = .2') )

plt.text(5,8,'this is a good plot,maybe...?', fontdict={'color' : 'r', 'size' : 30})

plt.show()
```
