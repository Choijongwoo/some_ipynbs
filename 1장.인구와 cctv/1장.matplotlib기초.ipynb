```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# colormap
plt.figure(figsize = (10,6))


X = np.linspace(0,10,10)
plt.plot(X,np.sin(X), label = "sin_wave", lw = 3, marker = 'o', markerfacecolor = 'purple', markersize = 15)
plt.plot(X,np.cos(X), "r",  label = "cos_wave", linestyle = 'dashed')
# plt.plot(X,np.tan(X), label = "tan_wave")



plt.xlabel("XLABEL", fontsize = 15)
plt.ylabel("YLABEL", fontsize = 15)
plt.title("SIN & COS", fontsize = 30)
plt.grid()
plt.legend()



plt.show()
```

```python
plt.figure(figsize = (10,6))


X = np.linspace(0,10,100)
plt.scatter(X,np.sin(X), label = "sin_wave", marker = '*', s = 100, c = np.sin(X))
plt.plot(X,np.cos(X), label = "cos_wave")
# plt.plot(X,np.tan(X), label = "tan_wave")


ax = plt.axes
plt.colorbar()
plt.xlabel("XLABEL", fontsize = 15,color = 'white')
plt.ylabel("YLABEL", fontsize = 15)
plt.title("SIN & COS", fontsize = 30)
plt.grid()
plt.legend()



plt.show()
```
